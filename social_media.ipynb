```python
import pandas as pd
import matplotlib.pyplot as plt

# -------------------------------
# STEP 1: CREATE CSV DATA
# -------------------------------
data = {
    "date": ["2024-01-01", "2024-01-02", "2024-01-03", "2024-01-04", "2024-01-05"],
    "video_id": ["V001", "V002", "V003", "V004", "V005"],
    "likes": [1200, 900, 1500, 800, 2000],
    "comments": [230, 180, 300, 140, 450],
    "shares": [150, 120, 210, 90, 320],
    "impressions": [15000, 12000, 18000, 10000, 22000],
    "watch_time_minutes": [4200, 3500, 5200, 2800, 6800]
}

df = pd.DataFrame(data)
df.to_csv("youtube_engagement.csv", index=False)

# -------------------------------
# STEP 2: LOAD DATA
# -------------------------------
df = pd.read_csv("youtube_engagement.csv")

# -------------------------------
# STEP 3: ENGAGEMENT METRIC
# -------------------------------
df["engagement_rate"] = (
    (df["likes"] + df["comments"] + df["shares"]) /
    df["impressions"]
) * 100

# -------------------------------
# STEP 4: KPI CALCULATION
# -------------------------------
total_likes = df["likes"].sum()
total_comments = df["comments"].sum()
total_shares = df["shares"].sum()
avg_engagement = round(df["engagement_rate"].mean(), 2)

print("===== SOCIAL MEDIA ENGAGEMENT DASHBOARD =====")
print(f"Total Likes       : {total_likes}")
print(f"Total Comments    : {total_comments}")
print(f"Total Shares      : {total_shares}")
print(f"Avg Engagement %  : {avg_engagement}")

# -------------------------------
# STEP 5: BEST PERFORMING CONTENT
# -------------------------------
best_video = df.loc[df["engagement_rate"].idxmax()]

print("\nBest Performing Video")
print("---------------------")
print(f"Video ID          : {best_video['video_id']}")
print(f"Engagement Rate % : {round(best_video['engagement_rate'],2)}")

# -------------------------------
# STEP 6: VISUALIZATION
# -------------------------------
plt.figure()
plt.plot(df["date"], df["engagement_rate"], marker='o')
plt.title("Engagement Rate Trend")
plt.xlabel("Date")
plt.ylabel("Engagement Rate (%)")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
